```python
# Base de dados criada utilizando MySQL Workbench + Wampserver
# OBS: é necessário executar o script SQL cliente_cidade.sql antes de executar esta célula


# Configuração da conexão com o banco de dados
db_config = {
    'user': 'root',
    'password': '',
    'host': 'localhost',
    'port': 3306,
    'database': 'cliente_cidade'
}

# Conectar ao banco de dados e executar a consulta
conn = mysql.connector.connect(**db_config)
cursor = conn.cursor()

# Selecionando os registros da tabela 'cliente' 
query = "SELECT * FROM cliente"
cursor.execute(query)

# Ler os resultados da consulta
result = cursor.fetchall()
column_names = [column[0] for column in cursor.description]
cliente_cidade_df = pd.DataFrame(result, columns=column_names)

print(cliente_cidade_df.head())

# Fechar a conexão
cursor.close()
conn.close()
```

# API IBGE
Código para utilizar a API do IBGE para recuperar os nomes de todos os municípios de Minas Gerais

Referência: https://servicodados.ibge.gov.br/api/docs/localidades

```python
url = "https://servicodados.ibge.gov.br/api/v1/localidades/estados/31/municipios"
response = requests.get(url)

if response.status_code == 200:
    municipios_data = response.json()
    municipios_nomes = [municipio['nome'] for municipio in municipios_data]
    print(municipios_nomes)
else:
    print(f"Erro ao acessar a API: {response.status_code}")
```
